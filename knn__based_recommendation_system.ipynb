# Here is the code to convert the invalid json to valid json and store the data as CSV(coma seperated Values)

```
import json
import ast
import csv
import pandas as pd
data_file = open('test3.csv', 'w') 
csv_writer = csv.writer(data_file)
with open('reviews.json') as json_file:      
    data1 = json_file.readlines()
data=list(data1)
k=[]
for i in data:
  h=ast.literal_eval(i)
  k.append(h)
 
  
# Counter variable used for writing  
# headers to the CSV file 
count = 0
  
for emp in k: 
    if count == 0: 
  
        # Writing headers of CSV file 
        header = emp.keys() 
        csv_writer.writerow(header) 
        count += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(emp.values()) 
  
data_file.close() 
```

## for epnion  data

```
from scipy.sparse import csr_matrix

book_features_df_matrix = csr_matrix(soc.values)

from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(book_features_df_matrix)
```

```
sc.shape
```

```
import numpy as np
query_index = np.random.choice(soc.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(soc.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```
soc.head()
```

```
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(soc.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, soc.index[indices.flatten()[i]], distances.flatten()[i]))
```
