# rename the columns with proper tags

```
rest=rest.rename(columns={'Unnamed: 0':'index','0':'Name','1':'Price','2':'Type of Cusine','3':'Timing ','4':'Address','5':'coordinates','6':'LandMark','7':'website','8':'Phone'})
```

```
rest['lati'] = rest['lati'].replace({"'":"","'":""})
```

```
rest[['lat','long']] = rest.coordinates.str.split(",",expand=True)
```

```
rest[['longit','longi']] = rest.long.str.split("]",expand=True)
```
